```python
import numpy as np
import networkx as nx
from extremalGraphRealizations import *

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
cases = ['tetrahedral_graph',
         'dodecahedral_graph',
         'octahedral_graph',
         'icosahedral_graph',
         'truncated_cube_graph',
         'truncated_tetrahedron_graph',
         'circular_ladder_graph',
         'cycle_graph_5',
         'bull_graph',
         'house_graph',
         'house_x_graph',
         'krackhardt_kite_graph',
         'cube_graph',
         'grid_2d_graph',
         'hexagonal_lattice_graph',
         'triangular_lattice_graph']
#g_n = cases[0]

g_n = 'test'
if g_n == 'test': G = nx.icosahedral_graph()

# Platonic
if g_n == 'tetrahedral_graph': G = nx.tetrahedral_graph()
if g_n == 'dodecahedral_graph': G = nx.dodecahedral_graph()
if g_n == 'octahedral_graph': G = nx.octahedral_graph()
if g_n == 'icosahedral_graph': G = nx.icosahedral_graph()
if g_n == 'truncated_cube_graph': G = nx.truncated_cube_graph()
if g_n == 'truncated_tetrahedron_graph': G = nx.truncated_tetrahedron_graph()
elif g_n == 'cube_graph': 
    G = nx.hypercube_graph(3) 
    G = nx.convert_node_labels_to_integers(G)
# cycles
elif g_n == 'circular_ladder_graph': G = nx.circular_ladder_graph(11)
elif g_n == 'cycle_graph_6': G = nx.cycle_graph(6)
# other
elif g_n == 'bull_graph': G = nx.bull_graph()
elif g_n == 'house_graph': G = nx.house_graph()
elif g_n == 'buckyball': G = nx.from_numpy_matrix(getBuckyAdj())    
# zero w value
elif g_n == 'house_x_graph': G = nx.house_x_graph()
elif g_n == 'krackhardt_kite_graph': G = nx.krackhardt_kite_graph()
# lattices
elif g_n == 'triangular_lattice_graph': 
    G = nx.triangular_lattice_graph(4,5)
    G = nx.convert_node_labels_to_integers(G)
elif g_n == 'grid_2d_graph':
    G = nx.grid_2d_graph(5,5)
    G = nx.convert_node_labels_to_integers(G)
elif g_n == 'hexagonal_lattice_graph':
    G = nx.hexagonal_lattice_graph(4,5)
    G = nx.convert_node_labels_to_integers(G)    
# higher multiplicity
elif g_n == 'petersen_graph': G = nx.petersen_graph() # mult = 5
elif g_n == 'heawood_graph': G = nx.heawood_graph() # mult = 6
elif g_n == 'pappus_graph': G = nx.pappus_graph() # mult = 6
# graph not found
else: print(f'graph_name {g_n} not known')
    
G.name = g_n  
phi=np.ones(G.number_of_edges())
#phi[0] = 1.5**2
spec_real = specRealization(G=G,phi=phi)

%matplotlib inline
if spec_real['mult'] == 1: 
    print('dimension is only one')
elif spec_real['mult'] == 2: 
    plotRealization2d(G,spec_real,save_tf=False)
    plt.show()
elif spec_real['mult'] == 3:
    plotRealization3d(G,spec_real,save_tf=False)
    plt.show()
else:
    print('dimension too large')  
```
